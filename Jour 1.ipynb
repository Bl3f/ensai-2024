````python
import json
from datetime import datetime, timedelta

from google.cloud import storage
from google.oauth2 import service_account

with open("ensai-2024-630e074aa45c.json") as source:
    info = json.load(source)

credentials = service_account.Credentials.from_service_account_info(info)

start = datetime(2024, 1, 1)
client = storage.Client(project="ensai-2024", credentials=credentials)
bucket = client.get_bucket('christophe-ensai-2024')

while start < datetime(2025, 1, 1):
    start += timedelta(days=1)
    folder_name = start.strftime("%Y-%m-%d")
    blob = bucket.blob(f"{folder_name}/")
    blob.upload_from_string('', content_type='application/x-www-form-urlencoded;charset=UTF-8')
````

````python
from datetime import datetime, timedelta
import io

import requests
from google.cloud import storage
import pandas as pd

date = (datetime.now() - timedelta(days=1)).strftime("%Y-%m-%d")
weather_url = f"https://public.opendatasoft.com/api/explore/v2.1/catalog/datasets/donnees-synop-essentielles-omm/records?where=date%20%3E%3D%20%22{date}%22&limit=20"

print(f"Downloading data for {date}...")

response = requests.get(weather_url)

with open("ensai-2024-630e074aa45c.json") as source:
    info = json.load(source)

credentials = service_account.Credentials.from_service_account_info(info)

client = storage.Client(project="ensai-2024", credentials=credentials)
bucket = client.get_bucket("weather-christophe")
blob = bucket.blob(f"{date}.parquet")

df = pd.DataFrame(response.json()["results"])
bytes_data = df.to_parquet()

blob.upload_from_string(bytes_data)
````
